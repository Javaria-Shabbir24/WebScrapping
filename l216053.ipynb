## Faculty of Lahore

- Loading the data 

```python
# lahore faculty website url
Lahore_Faculty_URL = "http://lhr.nu.edu.pk/faculty/" 
# request to get data
response = requests.get(Lahore_Faculty_URL) 
# print the response
print(response) 
```

- Creating soup object

```python
soup = BeautifulSoup(response.content, 'html.parser') 
```

```python
# holds the sets of each member data
lahore_faculty=[] 
href_value=''
# departments 
departments=["fsc","ee","cv","fsm","ss"] 
for x in departments:
    # iterating through each department
    faculty=soup.find(class_="container", id=x) 

    # extracting departments by getting the id associated with each class container
    department=faculty.get("id")

    # extracting the class associated with each faculty member
    members=faculty.findAll("div", class_="col-lg-3 col-md-4 col-sm-6 col-12")
    #iterating through each member data
    for m in members:
        member_id=0
        if m:
           # extracting id
           a_tag = m.find('a', href=True)
           # Extracting the last 4 digits from the href 
           if a_tag:
                href_value = a_tag['href']
                last_4_digits = href_value.split('/')[-1] 
                member_id= last_4_digits
    
           # extracting name
           name = m.find("h5",class_="text-center").text
           
           # extracting email
           email = m.find(class_= "mb-0 text-center").text.strip()
            
           # extracting image url
           url=m.find("img",class_="card-img-top rounded-circle mt-3 mb-0 d-block mx-auto").get("src")
           
           # extracting class possessing highest education and designation
           italic_texts = m.find_all(class_="small text-center font-italic")  
    
           for italic_text in italic_texts:
                br_tag = italic_text.find("br")  # Check if <br> exists
                
                if br_tag:
                    designation = br_tag.previous_sibling.strip() if br_tag.previous_sibling else None
                    x = br_tag.next_sibling.strip() if br_tag.next_sibling else None
                    if x and "HEC" in x:
                        education=x
                    else:
                        education=""
                        
                else:
                    designation = italic_text.get_text(strip=True)  # Get the whole text if no <br> exists
                    education = ""

            
# could not implement bcz of crashing the system
            
           # extracting complete education from the profile
           #edu_response= requests.get("http://lhr.nu.edu.pk"+href_value)
           #soup=BeautifulSoup(edu_response.content, "html.parser")
           #y=soup.find(class_="col-lg-8 col-md-6 col-sm-12 text-justify")
           #if y:
               #education=y.find("ul").find("li").text if y.find("ul").find("li") else ""
               
           # extracting designation and HECApprovedPhDSupervisor
           # x= soup.find(class_="small text-center font-italic")
           #br_tag=x.find("br")
           #if br_tag:
               #designation = br_tag.previous_sibling.strip()
               #HECApprovedPhDSupervisor= True
           #else:
               #designation = x.get_text(strip=True)
               #HECApprovedPhDSupervisor=False
          
           # setting the HECApprovedPhDSupervisor 
           HECApprovedPhDSupervisor= bool(education)
            
           extension=""
           # extracting extension number
           comments = m.find(string=lambda text: isinstance(text, Comment)) # returns a raw string
           # if any comment found extract extension number
           if comments:
               match = re.search(r"ext no:\s*(\d+)",comments)
               if match:
                   extension=match.group(1)
           # appending the member data 
           lahore_faculty.append({
                "ID": member_id,
                "Name": name,
                "Designation": designation,
                "HEC Approved PHD Supervisor": HECApprovedPhDSupervisor,
                "Highest Education": education,
                "Email": email,
                "Department":department,
                "Extension": extension,
                "ImageURL": url
           })
            
    
# creating dataframe
df_lahore=pd.DataFrame(lahore_faculty)
print(df_lahore)

# creating the csv file
df_lahore.to_csv("lhr.csv", index=True)
```

## Faculty of Karachi

- Loading the data

```python
# Karachi faculty website url
Karachi_Faculty_Url="https://khi.nu.edu.pk/faculty-php/"
# request to get details
response_Karachi=requests.get(Karachi_Faculty_Url)
# printing the response
print(response_Karachi)
```

- Creating soup object

```python
soupKhi= BeautifulSoup(response_Karachi.content,"html.parser")

# Extracting departments links from the navigation menu
menu = soupKhi.find(class_="menu-item menu-item-type-custom menu-item-object-custom current-menu-ancestor current-menu-parent menu-item-has-children menu-item-6256")
sub_menu=menu.find("ul",class_="sub-menu")
# Extract department links
department_links = []
if sub_menu:
    for li in sub_menu.findAll("li", class_="menu-item"):
        link= li.find("a",href=True)
        if link:
            #appending the departments links
            department_links.append(link["href"])
```

- Extracting faculty details

```python
# appending all the members of different departments
karachi_faculty=[] 

# iterating through each department
for dept in department_links:
    dept_url= requests.get(dept)
    soup= BeautifulSoup(dept_url.content,"html.parser")

    # main container possessing the details of HOD and other faculty members
    container= soup.find(class_="gdlr-core-page-builder-body clearfix")
    
    # extracting the department
    department= soup.find(class_="kingster-page-title").text
    faculty=container.find(class_="gdlr-core-personnel-item gdlr-core-item-pdb clearfix gdlr-core-left-align gdlr-core-personnel-item-style-grid-with-background gdlr-core-personnel-style-grid gdlr-core-with-background")
    for f in faculty:

        link=f.find(class_="gdlr-core-personnel-list-button gdlr-core-button")['href']
        
        # no employee id found
        id= ""
        
        # extracting Image urls
        url= f.find(class_="gdlr-core-personnel-list-image gdlr-core-media-image gdlr-core-hover-element gdlr-core-zoom-on-hover").find("img")["data-src"]
    
        # extracting name and education
        full_name=f.find("h3",class_="gdlr-core-personnel-list-title").text
        if "," in full_name:
            name, education=full_name.split(",")
        else:
            name=full_name
            
        # remove the on leave text
        if "(" in name and ")" in name:
            # Remove text inside parentheses
            name = re.sub(r"\s*\(.*?\)", "", name)
            
        # getting the education actual code
        # commented because the code is crashing repeatedly 
        # res=requests.get(link)
        # edu_soup=BeautifulSoup(res.content, "html.parser")
        # container=edu_soup.find("span", class_="gdlr-core-icon-list-content").text.strip() if edu_soup.find("span", class_="gdlr-core-icon-list-content") else education
        # education=container
        
        # setting the HECApprovedPhdSupervisor [ no information given about it ]
        # a phd supervisor is not always HEC Approved [not explicitly written]
        HECApprovedPhdSupervisor = False
    
        # extracting designation
        designation= f.find(class_="gdlr-core-personnel-list-position gdlr-core-info-font gdlr-core-skin-caption").text
    
        # extracting email
        email=f.find(class_="kingster-personnel-info-list kingster-type-email").text
    
        # extracting extension
        extension= f.find(class_="kingster-personnel-info-list kingster-type-phone").text
    
        #appending the data
        karachi_faculty.append({
                        "ID": id,
                        "Name": name,
                        "Designation": designation,
                        "HEC Approved PHD Supervisor": HECApprovedPhdSupervisor,
                        "Highest Education": education,
                        "Email": email,
                        "Department":department,
                        "Extension": extension,
                        "ImageURL": url
                   })
        # reset variables
        name=""
        education= ""
    
df_karachi=pd.DataFrame(karachi_faculty)
print(df_karachi)

# creating the csv file
df_karachi.to_csv("khi.csv", index=True)
```

## Faculty of Peshawar

- Loading the data

```python
#link for peshawar campus
url_psw="http://pwr.nu.edu.pk/"
#getting the response
response_psw= requests.get(url_psw)
print(response_psw)
```

- Creating the soup object

```python
soup_psw=BeautifulSoup(response_psw.content, "html.parser")
```

- Extracting links of all the departments

```python
# extracting the links of departments from navbar
dept_links_psw=[]
for li in soup_psw.find_all('li'):
    # if department link
    if 'Departments' in li.text:
        for sub_li in li.find_all('li'):
            link = sub_li.find('a')
            if link and 'href' in link.attrs:
                linkfaculty=link['href']
                # if faculty of department
                if 'faculty' in linkfaculty:
                    dept_links_psw.append(linkfaculty)
```

- Iterating through each department to get the data 

```python
# stores the sets of the faculty members details of different departments
psw_faculty=[]
for dept in dept_links_psw:
    id=''
    # response from the department
    response_dept=requests.get(url_psw+dept)
    dept_soup=BeautifulSoup(response_dept.content,"html.parser")
    
    # extracting the department name
    department= dept_soup.find(class_="main-content").find(class_="title text-white").text

    # extracting the faculty members 
    faculty=dept_soup.find(class_="faculty-list")
    faculty_members=faculty.findAll(class_="faculty-member")
    for m in faculty_members:
        
        # extracting id
        a_tag = m.find('a', class_='text-theme-color-2')
        if a_tag:
            href = a_tag.get('href')  
            match = re.search(r'id=(\d+)', href)  
            if match:
                id = match.group(1)

        # extracting image_url
        url=m.find('img')['src']

        # extracting name
        name= m.find("h2").text
        
        # extracting the designation, email, extension number, education, HECApprovedPhdSupervisor
        paragraphs= m.find(class_="faculty-details").findAll('p')
        if len(paragraphs)==3:
            designation=paragraphs[0].text
            email=paragraphs[1].text
            extension=paragraphs[2].text.strip().split('Ext. ')[-1].rstrip(')')
            education=""
            HECApprovedPhdSupervisor=False
        elif len(paragraphs)==4:
            designation=paragraphs[0].text
            education=paragraphs[1].text
            email=paragraphs[2].text
            extension=paragraphs[3].text.strip().split('Ext. ')[-1].rstrip(')')
            HECApprovedPhdSupervisor=True
            
        if (department=="HEC Approved Supervisors"):
            HECApprovedPhdSupervisor=True
            education= "HEC Approved Supervisors"
        
        #appending the data
        psw_faculty.append({
                        "ID": id,
                        "Name": name,
                        "Designation": designation,
                        "HEC Approved PHD Supervisor": HECApprovedPhdSupervisor,
                        "Highest Education": education,
                        "Email": email,
                        "Department":department,
                        "Extension": extension,
                        "ImageURL": url
                   })
       
    
df_psw=pd.DataFrame(psw_faculty)
print(df_psw)
# creating the csv file
df_psw.to_csv("psw.csv", index=True)


    
```

## Faculty of Faisalabad

- Loading the data

```python
# cfd url
cfd_url = 'https://cfd.nu.edu.pk/all-departments/'
# handling the 403 error
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36'
}
# response received
response_cfd = requests.get(cfd_url, headers=headers)
print(response_cfd.status_code)  
```

- Creating the soup object

```python
soup_cfd=BeautifulSoup(response_cfd.content, "html.parser")
```

- Extracting the department's links

```python
# stores the sets of faculty member details of different departments
cfd_faculty=[]
# stores the departments links
dept_list=[]
# extracting the departments links
navbar= soup_cfd.find(class_="menu-item menu-item-type-custom menu-item-object-custom current-menu-ancestor current-menu-parent menu-item-has-children menu-item-4508")
faculty_links=navbar.findAll("li")
for faculty in faculty_links:
    dept_link = faculty.find('a', href=lambda href: href and "#faculty" in href)
    if dept_link:
        dept_list.append(dept_link['href'])

# removing duplicate links
dept_list = list(set(dept_list))
```

- Iterating through each department's faculty members

```python
# no employee id is found
id=''
for dept in dept_list:
    response_dept= requests.get(dept,headers=headers)
    soup_dept= BeautifulSoup(response_dept.content,"html.parser")
    # extracting the department
    department= soup_dept.find(class_="breadcrumbs breadcrumbs-title-center").find("h2",class_="page-title").text
    # getting the details of all teachers
    teachers= soup_dept.find(class_="unitech-teachers").findAll(class_="unitech-teacher")
    for teacher in teachers:

        # extracting images_url
        url= teacher.find("img")["src"]

        teacher_detail= teacher.find(class_="unitech-teacher__inner")
        
        # extracting name
        name= teacher_detail.find("h4").text.strip()
        
        # extracting designation
        designation= teacher_detail.find("h6").text

        # extracting email and HECApprovedPhdSupervisor
        data= teacher.find("ul",class_="unitech-teacher-social-icon").find("li").findAll("p")
        # email extracted
        email= data[0].text
        # identifying HEC approved
        if(len(data)>1):
            # HECApprovedPhdSupervisor identifying
            if "HEC" in data[1].text:
                # setting hec approved true if found
                HECApprovedPhdSupervisor=bool(data[1].text)

        # extracting link for each person profile for remaining details
        teacher_url= teacher_detail.find("h4").find("a")["href"]

        # extracting the profile
        response_teacher_profile=requests.get(teacher_url,headers=headers)
        
        # creating the soup object
        teacher_soup=BeautifulSoup(response_teacher_profile.content,"html.parser")
        
        teacherAddress= teacher_soup.find(class_= "teacher__address").findAll("li")
        if teacherAddress[3]:
            extension=teacherAddress[3].find("span").next_sibling.text.strip() if teacherAddress[3].find("span") else ""

        # extracting highest education
        
        p_tag= teacher_soup.find(class_="htc__skill__container progress__bar--2").find("p")
        if p_tag:
            str=p_tag.find("strong")
            if str:
                education= p_tag.get_text(separator=",").strip()
            else:
                if p_tag.find("br"):
                    br_tag=p_tag.find("br")
                if br_tag:
                    education=br_tag.previous_sibling.strip()
                else:
                    education=p_tag.text.strip()

        else:
            ul_tag= teacher_soup.find(class_="htc__skill__container progress__bar--2").find("ul")
            if ul_tag:
                education=ul_tag.find("li").text

        #appending the data
        cfd_faculty.append({
                        "ID": id,
                        "Name": name,
                        "Designation": designation,
                        "HEC Approved PHD Supervisor": HECApprovedPhdSupervisor,
                        "Highest Education": education,
                        "Email": email,
                        "Department":department,
                        "Extension": extension,
                        "ImageURL": url
                   })
        # HECApprovedPhdSupervisor reset
        HECApprovedPhdSupervisor=False
       
    
df_cfd=pd.DataFrame(cfd_faculty)
print(df_cfd)
# creating the csv file
df_cfd.to_csv("cfd.csv", index=True)   
```

## Faculty of Islamabad

- Loading the data

```python
# islamabad faculty url
isb_url="http://isb.nu.edu.pk/Faculty/allfaculty/"

headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36',
    'X-Requested-With': 'XMLHttpRequest'}

# response received
response_isb = requests.get(isb_url, headers=headers)
print(response_isb.status_code)  
```

- Creating the soup object

```python
soup_isb=BeautifulSoup(response_isb.content, "html.parser")
```

- Extracting the faculty details

```python
# stores the sets of members details
isl_faculty=[]
#extracting the departments
container=soup_isb.findAll(class_="container")
# url to get details of all employees of a department
employees_details_url="http://isb.nu.edu.pk/Faculty/GetAllEmp"
x=0 # variable for counter
for c in container:
    x=x+1
    # ensuring no duplicates occur
    if x<5:
        if x==1:
            # id for computing department
            C_id=301 
        elif x==2:
            # id for Electrical Engineering department
            C_id=302
        elif x==3:
            # id for Management Sciences department
            C_id=303
        elif x==4:
            #id for Science & Humanities
            C_id=313
            
        # ensure the departments
        department = c.get("id")
        # extracting the department name
        if department:
            department=c.find(class_="animated fadeInLeft").text
            
        # sending the request to get the faculty details page
        faculty_response = requests.get(employees_details_url, params={'id': C_id}, headers=headers)
        
        faculty_members = faculty_response.json()
        for m in faculty_members:
            # extracting id
            id = m["Emp_ID"]
            
            # extracting name 
            name = m["Name"].strip()
    
            # extracting email
            email = m["Email"]
    
            # extracting image url
            url= m["ImagePath"]
            HECApprovedPhdSupervisor=False
            # HECApprovedPhdSupervisor
            HECApprovedPhdSupervisor = m['hec_supervisor']
            if HECApprovedPhdSupervisor==None:
                HECApprovedPhdSupervisor=False
            
            if HECApprovedPhdSupervisor:
                education="HEC Approved Phd Supervisor"
    
            # extracting extension
            extension =m['Extension']
    
            # extracting designation
            designation =m["Designation_Title"]

            #appending the data
            isl_faculty.append({
                        "ID": id,
                        "Name": name,
                        "Designation": designation,
                        "HEC Approved PHD Supervisor": HECApprovedPhdSupervisor,
                        "Highest Education": education,
                        "Email": email,
                        "Department":department,
                        "Extension": extension,
                        "ImageURL": url
                   })
            education=""

df_isl=pd.DataFrame(isl_faculty)
print(df_isl)
# creating the csv file
df_isl.to_csv("isl.csv", index=True)  

  
```

## Concatenation of DataFrames

```python
# Concatenating while resetting the index
fast_faculty = pd.concat([df_lahore, df_karachi, df_psw, df_cfd, df_isl], ignore_index=True)

# Displaying the combined DataFrame
print(fast_faculty)

# CSV file 
fast_faculty.to_csv("fast_faculty.csv", index=True)
```
